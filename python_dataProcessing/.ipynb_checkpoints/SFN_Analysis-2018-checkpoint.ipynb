```python
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import os
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Set plot style
custom_style = {'figure.facecolor':'.4',
                "axes.facecolor":".4",
                'axes.edgecolor':'.4',
                "axes.labelcolor":"white",
                "axes.grid":True,
                'grid.color':'white',
                "text.color":"white",
                "patch_edgecolor":'white',
                "xtick.color":"white",
                "ytick.color":"white",
                'axes.edgecolor':'black',
               'legend.frameon':True}
sns.set_style('white',rc=custom_style)
sns.set_context('poster')
col_palette = sns.color_palette("husl")
def addTitle(fig=None,title='',adj=0.85):
    plt.subplots_adjust(top=adj)
    fig.fig.suptitle(title)
```

```python
traceList = []
specList = []
bandList = []
project_path = '/home/roshan/Projects/'
for dirpath,dirnames,files in os.walk(project_path):
    for name in files:
        if name.endswith('deltaTraceMatrix.csv'):
            traceList.append(os.path.join(dirpath+'/'+name))
        if name.endswith('sleepSpecMatrix.csv'):
            specList.append(os.path.join(dirpath+'/'+name))
        if name.endswith('sleepBandPowerMatrix.csv'):
            bandList.append(os.path.join(dirpath+'/'+name))

```

```python
traceList
```

```python

traceMat = pd.DataFrame()
specMat = pd.DataFrame()
bandMat = pd.DataFrame()
for rf in traceList:
    tmp = pd.read_csv(rf)
    traceMat = traceMat.append(tmp)
for rf in specList:
    tmp = pd.read_csv(rf)
    specMat = specMat.append(tmp)
for rf in bandList:
    tmp = pd.read_csv(rf)
    bandMat = bandMat.append(tmp)
traceMat = traceMat.reset_index(drop=True)
bandMat = bandMat.reset_index(drop=True)
specMat = specMat.reset_index(drop=True)
print('Data Loaded!')
```

```python
traceMat['epoch_timestamp'] = pd.to_datetime(traceMat['epoch_time'],unit='s',errors='raise')
#traceMat.loc[traceMat['epoch_timestamp'].isnull(),'epoch_timestamp']=pd.to_datetime(traceMat[traceMat['epoch_timestamp'].isnull()].epoch_time,unit='s')
```

```python
# Change Sleep to Baseline
idx = traceMat.epoch_type=='Sleep'
traceMat.loc[idx,'epoch_type']='Baseline'
idx = specMat.epoch_type=='Sleep'
specMat.loc[idx,'epoch_type']='Baseline'
idx = bandMat.epoch_type=='Sleep'
bandMat.loc[idx,'epoch_type']='Baseline'
```

```python
idx = traceMat.sleep_state!=6
idx = idx & (traceMat.day!=8)
idx = idx & (traceMat.gender=='M')
newTrace = traceMat.loc[idx,:].groupby(['animal','genotype','day','epoch_type','tetrode'])['epoch_timestamp','epoch_time','age','sleep_state','STFT_delta_power','CWT_delta_power'].resample('1T',on='epoch_timestamp').mean()
newTrace = newTrace.reset_index()
newTrace.head()
```

```python
def trace_plot(data,method,group,group_order=None,nBoot=1000):
    yCol = method+'_delta_power'
    print(yCol)
    g = sns.FacetGrid(data,hue='epoch_type',row=group,col='epoch_type',
                      margin_titles=True,sharex=False,height=5,aspect=2,
                      row_order=group_order,palette=col_palette,col_order=['Baseline','Saline','Ketamine'])
    g = g.map(sns.lineplot,'epoch_time',yCol,n_boot=nBoot)
    plt.subplots_adjust(top=0.9)
    g.fig.suptitle(method+' Delta Power')
    [plt.setp(ax.texts, text="") for ax in g.axes.flat]
    g.set_titles(row_template='{row_name}', col_template='{col_name}')
    g.set_axis_labels('','')
    ax = g.axes.flat
    g.fig.text(0,.5,'Normalized '+method+' Power (dB)',rotation=90,va='center',fontsize=24)
    g.fig.text(0.5,0,'Frequency (Hz)',ha='center',fontsize=24)
    g.despine()
```

```python
def logMean(x):
    tmp = np.mean(x)
    return 10*np.log10(tmp)
def spec_plot(data,method,group,asp=1.33,group_order=None,nBoot=1000):
    yCol = method+'_Power'
    if len(data[group].unique())>4:
        legend_y = 1.2
        cWrap=4
    else:
        legend_y = 1.05
        cWrap=2
    g = sns.FacetGrid(data=data,col=group,hue='state_name',
                      margin_titles=True,hue_order=('NREM','Rest','Active'),
                      palette=col_palette,col_order=group_order,height=6,aspect=asp,col_wrap=cWrap)
    g = g.map(sns.lineplot,'frequency',yCol,sort=True,n_boot=nBoot,estimator=logMean)
    plt.subplots_adjust(top=0.8)
    g.fig.suptitle('Mean '+method+' Spectra')
    [plt.setp(ax.texts, text="") for ax in g.axes.flat]
    g.set_titles(row_template='{row_name}', col_template='{col_name}')
    g.set_axis_labels('Frequency (Hz)',method+' Power (dB)')
    legend = plt.legend(loc='upper right', bbox_to_anchor=(1.1, legend_y),
                        ncol=1, fancybox=True, shadow=True,fontsize=18,title='states'.capitalize())
    g.despine()
```
